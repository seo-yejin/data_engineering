```python
import pandas as pd
import numpy as np
from pandas import datetime
from matplotlib import pyplot as plt
```

```python
def parser(x):
    return datetime.strptime(x, '%Y-%m-%d %H:%M:%S')

input_file = './data/AirQualityUCI_refined.csv'

df = pd.read_csv(input_file,
                 index_col=[0],
                 parse_dates=[0],
                 date_parser=parser)

df.head()
```

```python
df['PT08.S3(NOx)'].plot()
```

```python
co = df['PT08.S3(NOx)'].copy().interpolate()
```

```python
plt.plot(df['PT08.S3(NOx)'], label='original', zorder=2)
plt.plot(co, label='linear interpolation', zorder=1)
plt.legend(loc='best')
```

```python
plt.boxplot(co)
plt.title("Detecting outliers using Boxplot")
```

```python
q1 = co.quantile(0.25)
median = co.quantile(0.5)
q3 = co.quantile(0.75)
print(q1, median, q3)
```

```python
iqr = q3 - q1
upper_fence = q3 + 1.5 * iqr
lower_fence = q1 - 1.5 * iqr
print(iqr, upper_fence, lower_fence)
```

```python
outliers = co.loc[(co > upper_fence) | (co < lower_fence)]
print(outliers)
```

```python
mask = co.index.isin(outliers.index)
```

```python
co[~mask]
```

```python
plt.plot(co[~mask], label='normal', color='blue',
    marker='o', markersize=3, linestyle='None')
plt.plot(outliers, label='outliers', color='red',
    marker='x', markersize=3, linestyle='None')
plt.legend(loc='best')
```

```python
co_refined = co.copy()
co_refined[mask] = np.nan
```

```python
co_refined[mask]
```

```python
co_refined.interpolate(inplace=True)
```

```python
co_refined.plot()
```

```python
import seaborn as sns
```

```python
sns.distplot(co)
```

```python
mean = np.mean(co)
std = np.std(co)
print(mean, std)
```

```python
outliers = []
thres = 3

for i in co:
    z_score = (i - mean) / std
    if (np.abs(z_score) > thres):
        print(z_score)
        outliers.append(i)
```

```python
outliers = co.loc[np.abs((co - mean) / std) > 3]
outliers
```

```python
mask = co.index.isin(outliers.index)
```

```python
sns.distplot(co, axlabel='PT08.S3(NOx)', label='original')
sns.distplot(co[~mask], label='outliers removed')
plt.legend(loc='best')
```
